```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
width, height = 640, 320
img = 255 * np.ones(shape=[height, width, 3], dtype=np.uint8)

# Add background noise
#row,col,ch = img.shape
##mean = 128
#sigma = 32
#gauss = np.random.normal(mean,sigma,(row,col,ch))
#gauss = gauss.reshape(row,col,ch)
#img = gauss.astype(int)

for i in range(6):
    if np.random.rand() > 0.5:
        pt1_x = np.random.randint(low=0, high=width)
        pt1_y = np.random.randint(low=0, high=height)
        pt2_x = np.random.randint(low=0, high=width)
        pt2_y = np.random.randint(low=0, high=height)
        clr1 = np.random.randint(low=0, high=255)
        clr2 = np.random.randint(low=0, high=255)
        clr3 = np.random.randint(low=0, high=255)
        rnd_clr = (clr1, clr2, clr3)
        rnd_thickness = np.random.randint(low=5, high=15)
        res = cv2.rectangle(img, pt1=(pt1_x, pt1_y), pt2=(pt2_x, pt2_y), color=rnd_clr, thickness=rnd_thickness)
    if np.random.rand() > 0.5:
        pt1_x = np.random.randint(low=width//4, high=3*width//4)
        pt1_y = np.random.randint(low=height//4, high=3*height//4)
        min_dim = min(width, height)
        radius = np.random.randint(low=0, high=min_dim//4)
        clr1 = np.random.randint(low=0, high=255)
        clr2 = np.random.randint(low=0, high=255)
        clr3 = np.random.randint(low=0, high=255)
        rnd_clr = (clr1, clr2, clr3)
        rnd_thickness = np.random.randint(low=5, high=15)
        res = cv2.circle(img, center=(pt1_x,pt1_y), radius=100, color=rnd_clr, thickness=rnd_thickness)
    if np.random.rand() > 0.5:
        pt1_x = np.random.randint(low=0, high=width)
        pt1_y = np.random.randint(low=0, high=height)
        pt2_x = np.random.randint(low=0, high=width)
        pt2_y = np.random.randint(low=0, high=height)
        clr1 = np.random.randint(low=0, high=255)
        clr2 = np.random.randint(low=0, high=255)
        clr3 = np.random.randint(low=0, high=255)
        rnd_clr = (clr1, clr2, clr3)
        rnd_thickness = np.random.randint(low=5, high=15)
        res = cv2.line(img, pt1=(pt1_x, pt1_y), pt2=(pt2_x, pt2_y), color=rnd_clr, thickness=rnd_thickness)
```
